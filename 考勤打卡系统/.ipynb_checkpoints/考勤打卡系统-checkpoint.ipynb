```python
#  三元组损失函数

def triplet_loss(y_true, y_pred, alpha = 0.2):
    
    anchor, positive, negative = y_pred[0], y_pred[1], y_pred[2]
    
    # Step 1: 计算锚点和正点之间的距离(编码)
    pos_dist = tf.square(anchor-positive)

    # Step 2: 计算锚点和负数之间的(编码)距离
    neg_dist = tf.square(anchor-negative)

    # Step 3: 减去之前的两个距离，加上alpha.
    basic_loss = tf.reduce_sum(pos_dist-neg_dist)+alpha
    
    # Step 4: 取basic_loss和0.0的最大值。总结训练实例.
    loss = tf.reduce_sum(tf.maximum(basic_loss,0.))
    
    return loss
```

```python
with tf.Session() as test:
    tf.set_random_seed(1)
    y_true = (None, None, None)
    y_pred = (tf.random_normal([3, 128], mean=6, stddev=0.1, seed = 1),
              tf.random_normal([3, 128], mean=1, stddev=1, seed = 1),
              tf.random_normal([3, 128], mean=3, stddev=4, seed = 1))
    loss = triplet_loss(y_true, y_pred)
    print("loss = " + str(loss.eval()))
```

```python
database = {}
database["liujiyong"] = img_to_encoding("images/ljy.jpg", FRmodel)
database['cuishu'] = img_to_encoding('images/cuishu.jpg',FRmodel)

print(database)
```

```python
# GRADED FUNCTION: verify

def verify(image_path, identity, database, model):
    
    # Step 1: 转换为图形编码矩阵
    encoding = img_to_encoding(image_path, FRmodel)
    
    # Step 2: 计算距离
    dist = np.linalg.norm(encoding-database[identity])
    
    # Step 3: 如果dist < 0.7，则打开门，否则不打开
    if dist<0.7:
        print("It's " + str(identity) + ", welcome home!")
        door_open = 1
    else:
        print("It's not " + str(identity) + ", please go away")
        door_open = 0
        
    return dist, door_open
```

```python
verify("images/ljy.jpg", "liujiyong", database, FRmodel)
verify("images/cuishu2.jpg", "cuishu", database, FRmodel)
```

```python

def who_is_it(image_path, database, model):
    
    ## Step 1: 转换图形编码矩阵
    encoding = img_to_encoding(image_path, model)
    ## Step 2:找到最接近的编码##
    min_dist = 100
    
    for (name, db_enc) in database.items():
        dist = np.linalg.norm(encoding-database[name])
        
        if dist<min_dist:
            min_dist = dist
            identity = name

    if min_dist > 0.7:
        print("不在数据集里.")
    else:
        print ("这是 " + str(identity) + ", 距离为: " + str(min_dist))
        
    return min_dist, identity
```
